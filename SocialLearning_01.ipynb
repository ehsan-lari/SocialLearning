```python
import torch
```

```python
T_MAX = 100
MU_0 = torch.Tensor([1/3, 1/3, 1/3])
MU = MU_0
N_STATES = len(MU_0)
MU_TEMP = torch.zeros(N_STATES)
V = torch.Tensor([1/2, 1, 3/2])
MU_ALL = torch.zeros(T_MAX, N_STATES)
```

```python
for t in range(T_MAX):
    MU_ALL[t, :] = MU
    x = torch.randn(1) + torch.Tensor([1.6])
    for n in range(N_STATES):
        MU_TEMP[n] = MU[n] * torch.exp( - ( x - V[n] )**2 / 2 ) / torch.sum( MU*torch.exp( - ( x*torch.ones(N_STATES) - V )**2 /2 ) )
    MU = MU_TEMP
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 6))
plt.plot(MU_ALL.detach().numpy())
plt.grid(axis='both')
plt.xlabel('t')
plt.ylabel('MU')
plt.legend(['mu1', 'mu2', 'mu3'])
plt.show()
```
